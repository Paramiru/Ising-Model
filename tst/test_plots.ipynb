## Test observables obtained from simulation
Run all the cells at once to obtain the following plots with respect to the temperature of the simulation:
1. Energy using Glauber dynamics
2. Heat Capacity using Glauber dynamics
3. Magnetisation using Glauber dynamics
4. Susceptibility using Glauber dynamics
5. Energy using Kawasaki dynamics
6. Heat Capacity using Kawasaki dynamics

Note: There is no point on showing the magnetisation and/or susceptibility with Kawasaki dynamics. Such algorithm does not change the number of spins in the system, but its location. Therefore the magnetisation will be 0 for the whole simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.style.use('seaborn-whitegrid')
```

```python
glauber_energy = np.loadtxt('../output_files/glauber50_energy,heat,heat_error.txt', delimiter=' ')
glauber_mgn = np.loadtxt('../output_files/glauber50_mgn,chi,chi_error.txt', delimiter=' ')
kawasaki_energy = np.loadtxt('../output_files/kawasaki50_energy,heat,heat_error.txt',  delimiter=' ')
ts = np.arange(1.0, 3.1, step=0.2)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(25,20), sharex=True)

axs[0,0].plot(glauber_energy[:,0], glauber_energy[:,1])
axs[0,0].scatter(glauber_energy[:,0], glauber_energy[:,1])

axs[0,0].set_title('Energy vs Temperature')
axs[0,0].set_ylabel('Energy')
axs[0,0].set_xticks(ts)
axs[0,0].set_ylim([-5000, -2000])
axs[0,0].set_xlim([1.0, 3.0])

axs[0,1].plot(glauber_energy[:,0], glauber_energy[:,2])
axs[0,1].errorbar(glauber_energy[:,0], glauber_energy[:,2], glauber_energy[:,3], 
    fmt='x', ecolor = 'mediumslateblue', color='m', capsize=5)
axs[0,1].set_title('Heat Capacity vs Temperature')
axs[0,1].set_ylabel('Heat Capacity')

axs[1,0].plot(glauber_mgn[:,0], glauber_mgn[:,1])
axs[1,0].scatter(glauber_mgn[:,0], glauber_mgn[:,1])
axs[1,0].set_title('Magnetisation vs Temperature')
axs[1,0].set_xlabel('Temperature (K)')
axs[1,0].set_ylabel('Magnetisation')

axs[1,1].errorbar(glauber_mgn[:,0], glauber_mgn[:,2], glauber_mgn[:,3], fmt='x',
             ecolor = 'mediumslateblue', color='m', capsize=5)
axs[1,1].plot(glauber_mgn[:,0], glauber_mgn[:,2])
axs[1,1].set_title('Susceptibility vs Temperature')
axs[1,1].set_xlabel('Temperature (K)')
axs[1,1].set_ylabel('Susceptibility')

fig.suptitle('Observables from Glauber Dynamics', fontsize=24)
fig.tight_layout()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(25,12))

ax0.set_title('Energy vs Temperature')
ax0.set_xlabel('Temperature (K)')
ax0.set_ylabel('Energy')
ax0.set_xticks(ts)
ax0.set_ylim([-5000, -2000])
ax0.set_xlim([1.0, 3.0])
ax0.plot(kawasaki_energy[:,0], kawasaki_energy[:,1])
ax0.scatter(kawasaki_energy[:,0], kawasaki_energy[:,1])

ax1.plot(kawasaki_energy[:,0], kawasaki_energy[:,2])
ax1.errorbar(kawasaki_energy[:,0], kawasaki_energy[:,2], kawasaki_energy[:,3], fmt='x',
             ecolor = 'mediumslateblue', color='m', capsize=5)
ax1.set_title('Heat Capacity vs Temperature')
ax1.set_xlabel('Temperature (K)')
ax1.set_ylabel('Heat Capacity')

fig.suptitle('Observables from Kawasaki Dynamics', fontsize=24)
fig.tight_layout()
```

```python
# Plot Energy vs Temperature - Glauber
fig_energy = plt.figure(figsize=(20,12))
plt.title('Energy vs Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Energy')
plt.xticks(ts)
plt.scatter(glauber_energy[:,0], glauber_energy[:,1])
plt.plot(glauber_energy[:,0], glauber_energy[:,1])
```
